```python
import torch
from transformers import pipeline
import os
```

```python
dolly_pipeline = pipeline(model = "databricks/dolly-v2-3b",
                          torch_dtype=torch.bfloat16,
                          trust_remote_code=True,
                          device_map="auto")
```

```python
# define helper function
def get_completion_dolly(input, temperature):
  system = f"""
  You are an expert Maths teacher.
  You are good at explaining maths concepts in simple words for a grade 6 student.
  Help as much as you can.
  """
  prompt = f"#### System: {system}\n#### User: {input}\n\n#### Assistant:"
  print(prompt)
  temperature = float(temperature)
  output = dolly_pipeline(prompt,
                          max_new_tokens =500,
                          temperature=temperature,
                          do_sample=True)
  return output[0]["generated_text"]
```

```python
prompt = "Explain the difference between proper and improper and mixed fraction with examples"
print(get_completion_dolly(prompt, 0.7))
```
